```python
import pandas as pd
import numpy as np

df = pd.read_csv('updated_with_learnerType.csv')

# Inspect Aptitude_Score to decide on a threshold
print(df['Aptitude_Score'].describe())

# Check for nulls
print("Nulls in Aptitude_Score:", df['Aptitude_Score'].isnull().sum())
```

```python
# Calculate threshold (mean)
threshold = df['Aptitude_Score'].mean()

# Define functions for update
def get_placement_domain(score, threshold):
    if pd.isna(score):
        return 'non-tech'
    if score > threshold:
        return 'tech'
    else:
        return 'non-tech'

def get_role(domain):
    if domain == 'tech':
        return 'Software Engineer'
    else:
        return 'Sales Executive'

# Apply updates
df['Placement_domain'] = df['Aptitude_Score'].apply(lambda x: get_placement_domain(x, threshold))
df['Role'] = df['Placement_domain'].apply(get_role)

# Save the updated dataframe
df.to_csv('updated_placement_by_aptitude.csv', index=False)

# Show sample results
print("Threshold used:", threshold)
print(df[['Aptitude_Score', 'Placement_domain', 'Role']].head(15))
print(df['Placement_domain'].value_counts())
print(df['Role'].value_counts())
```
